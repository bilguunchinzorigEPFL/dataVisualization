These are the question I'm intending to answer
* Is there a trend for the number of patents applications during the months of 2012?
* Which countries have the most assigned patents?
* What are the most popular technology fields for patents assigned in months of 2012?
* What's the most frequent technology field by inventor's country?

```python
#Basic imports
import requests
from bs4 import BeautifulSoup
import pandas as pd
%matplotlib inline
import seaborn as sbn
```

```python
#Here I'm dividing query into few main parts according to the API documentation
u_home='http://www.patentsview.org/api/patents/query?'
u_q='q='+'{"_and":[{"_eq":{"patent_date":"2012-1-3"}}]}' #,{"_eq":{"wipo_field_id":"36"}}
u_f='f='+'["patent_number","patent_date","wipo_field_title"]'
u_s='s='+'[]'
u_o='o={"page":1,"per_page":10000}'
#testing
r = requests.get(u_home+u_q+'&'+u_f+'&'+u_o)
print(u_home+u_q+'&'+u_f)
data = r.json()
```

```python
#pd.DataFrame(data['patents'])
data["patents"]
```

```python
#Now trying to get number of patents in each category
total=[]
category=[]
u_f='f='+'["patent_number","patent_date"]'
#in this case since we only need number of data we only need total patent count parameter from request
u_o='o={"page":1,"per_page":1}'
for month in range(1,13):
    #the reason why, I'm additionally getting total number of data is one patent can be in multiple fields and,
    #thus we can't acquire true total number of patents just by summing category data
    u_q='q='+'{"_and":[{"_gte":{"patent_date":"2012-'+str(month)+'-1"}},{"_lte":{"patent_date":"2012-'+str(month)+'-31"}}]}'
    r = requests.get(u_home+u_q+'&'+u_f+'&'+u_o)
    data = r.json()
    total.append(data["total_patent_count"])
    tmp=[0]*37
    #there is in total 37 technological category of WIPO
    for field in range(0,37):
        u_q='q='+'{"_and":[{"_gte":{"patent_date":"2012-'+str(month)+'-1"}},{"_lte":{"patent_date":"2012-'+str(month)+'-31"}},{"_eq":{"wipo_field_id":"'+str(field)+'"}}]}'
        r = requests.get(u_home+u_q+'&'+u_f+'&'+u_o)
        data = r.json()
        tmp[field]+=data["total_patent_count"]
    category.append(tmp)
    print(month)
```

```python
#total number of patents in each months
total
```

```python
#Now this time I'm trying to get patents and their citation relations
citations=[]
u_f='f='+'["patent_number","cited_patent_number","citedby_patent_number"]'
u_o='o={"page":1,"per_page":10000}'
for month in range(1,13):
    #Since we can get at maximum 10000 patents and in general there is 30000 patents in each month, I'm dividing month
    #by 5 intervals with 6 days (so that 5*6=30 days) to acquire the data
    for sixth in range(0,4):
        u_q='q='+'{"_and":[{"_gte":{"patent_date":"2012-'+str(month)+'-'+str(sixth*6)+'"}},{"_lte":{"patent_date":"2012-'+str(month)+'-'+str((sixth+1)*6)+'"}}]}'
        r = requests.get(u_home+u_q+'&'+u_f+'&'+u_o)
        data = r.json()
        citations.append(data["patents"])
    print(month)
```
